```python
df_country= pd.read_excel('Country-Code.xlsx')
```

```python
df_country.head()
```

```python
final_df=pd.merge(df,df_country, on='Country Code', how='left')
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
country_names= final_df.Country.value_counts().index
```

```python
final_df.Country.value_counts().index
```

```python
country_val= final_df.Country.value_counts().values
```

### Pie Chart

```python
plt.pie(country_val[:3], labels=country_names[:3], autopct='%1.2f%%')
```

### Observation: Nearly 95% of the orders of Zomato are from India, followed by USA and UK

```python
ratings= final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'rating count'})
```

```python
ratings
```

### 1. When Rating>=4.5 Excellent, Rating<=4.5 nd >=4.0, very good, Rating>=3.5 and <=3.9, good and so on....

### 2. Too many orders are going unrated

```python
ratings.head()
```

```python
plt.rcParams['figure.figsize']=(12,6)
sb.barplot(x="Aggregate rating",y="rating count", data= ratings, hue="Rating color", palette=['white','red','orange','yellow','green'])
```

```python
sb.countplot(x="Rating color", data=ratings,palette=['white','red','orange','yellow','green'])
```
